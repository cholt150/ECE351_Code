```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 14}) # set font size in plots
steps = 1e-2
t = np.arange(0,5+steps,steps)

# --- User-Defined Function --- #
# Create the output `y(t)` using a for loop and if/else statements
def example1(t):
    y = np.zeros((len(t),1)) # initialize `y` as a numpy array (of zeros)

    for i in range(len(t)):
        if i < (len(t)+1) / 3:
            y[i] = t[i]**2
        else:
             y[i] = np.sin(5*t[i]) + 2
    return y

y = example1(t) # function call using the user-defined function, shown in the above cell
myFigSize = (10,8)
plt.figure(figsize=myFigSize)
plt.subplot(2,1,1)
plt.plot(t,y)
plt.grid(True)
plt.ylabel('y(t) with Good Resolution')
plt.title('Background - Illustration of for Loops and if/else Statements')
# `t` defined with poor resolution
t = np.arange(0,5+0.25,0.25)
y = example1(t)
plt.subplot(2,1,2)
plt.plot(t,y)
plt.grid(True)
plt.ylabel('y(t) with Poor Resolution')
plt.xlabel('t')
plt.show() # this must be used for each figure
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 14}) # set font size in plots
steps = 1e-2
t = np.arange(0,10+steps,steps)
def func1(t):
    y = np.zeros((len(t),1))
    for i in range(len(t)):
        y[i] = np.cos(t[i])
    return y

f = func1(t)
myFigSize = (10,8)
plt.figure(figsize=myFigSize)
plt.subplot(2,1,1)
plt.plot(t,f)
plt.grid(True)
plt.ylabel('f(t) as a cosine')
plt.title('For-loop Cosine')
plt.show()
```

$$y=r(t)-r(t-3)+5u(t-3)-2u(t-6)-2r(t-6)$$

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(-15,20+steps,steps) #DEFINE TIME ARRAY

#RAMP FUNCTION
def r(t):
    y = np.zeros((len(t),1))
    for i in range(len(t)):
        if t[i] > 0:
            y[i] = t[i]
        else:
            y[i] = 0
    return y

#STEP FUNCTION
def u(t):
    y = np.zeros((len(t),1))
    for i in range(len(t)):
        if t[i] < 0:
            y[i] = 0
        else:
            y[i] = 1
    return y

#COMPOUND FUNCTION
def function(t):
    f = r(t) - r(t-3) + 5*u(t-3) - 2*u(t-6) -2*r(t-6)
    return f

f1 = function(-t)
f2 = function(t-4)
f3 = function(-t-4)
f4 = function(t/2)
f5 = function(2*t)

#DERIVATIVE MATH
dt = np.diff(t)
f6 = np.diff(function(t),axis=0)/dt

#Ramp Plot
plt.figure(figsize=(10,8))
plt.plot(t,r(t))
plt.grid(True)
plt.ylabel('Ramp Function')
plt.xlabel('t')
plt.axis([-2,2,-2,2])
plt.show()

#Step Plot
plt.figure(figsize=(10,8))
plt.plot(t,u(t))
plt.grid(True)
plt.ylabel('Step Function')
plt.xlabel('t')
plt.axis([-2,2,-2,2])
plt.show()

#Compound Plot
plt.figure(figsize=(10,8))
plt.plot(t,function(t))
plt.grid(True)
plt.ylabel('Compound Function')
plt.xlabel('t')
plt.axis([-1,9,-1,9])
plt.show()

#Reversal Plot
myFigSize = (15,25)
plt.figure(figsize=myFigSize)
plt.subplot(4,1,1)
plt.plot(t,function(t),label='f(t)')
plt.plot(t,f1,label='f(-t)')
plt.axis([-10,10,-2,9])
plt.grid(True)
plt.ylabel('f(t)')
plt.legend(loc='upper right')
plt.title('Time Reversal')

#Time Shift Plot
plt.subplot(4,1,2)
plt.plot(t,f2,label='f(t-4)')
plt.plot(t,f3,label='f(-t-4)')
plt.axis([-15,15,-2,9])
plt.grid(True)
plt.ylabel('f(t)')
plt.legend(loc='upper right')
plt.title('Time Shift Operations')

#Time Scale Plot
plt.subplot(4,1,3)
plt.plot(t,f4,label='f(.5t)')
plt.plot(t,f5,label='f(2t)')
plt.axis([0,20,-2,9])
plt.grid(True)
plt.ylabel('f(t)')
plt.legend(loc='upper right')
plt.title('Time scale Operations')

#Derivative Plot
plt.subplot(4,1,4)
plt.plot(t[range(len(f6))],f6)
plt.plot(t,function(t),label='f(t)')
plt.axis([-1,10,-4,10])
plt.grid(True)
plt.ylabel('f(t)')
plt.legend(loc='upper right')
plt.title('Derivative of f(t)')

plt.show()
```
